```python
import requests
import pandas as pd
from IPython.display import display
crtfc_key='d6a0116f591ed1ddb0398eb134e6813ddde6c015'

page_no=1
page_count=100
bgn_de='20230930'
end_de='20231230'

results_all=pd.DataFrame()
while(True):
    url = 'https://opendart.fss.or.kr/api/list.json'
    params = {
        'crtfc_key': crtfc_key,
        'page_no' : str(page_no),
        'page_count' : str(page_count),
        'bgn_de' : bgn_de,
        'end_de' : end_de,
    }

    #결과를 json형태로 저장
    results = requests.get(url, params=params).json()
    #공시정보가 있는부분만 DataFrame으로 저장
    results_df = pd.DataFrame(results['list'])
    results_all = pd.concat([results_all,results_df])

    total_page=results['total_page']

    if page_no==total_page:
        break

    page_no=page_no+1

keyword='임원ㆍ주요주주'
result_elestock=results_all.loc[results_all['report_nm'].str.contains(keyword)]
print(result_elestock.shape)
print(result_elestock)
```

```python
import pandas as pd
import requests
import xml.etree.ElementTree as et
from io import BytesIO
from zipfile import ZipFile

### OPENDART 개발 가이드
# https://opendart.fss.or.kr/guide/main.do?apiGrpCd=DS001
api_key = 'd6a0116f591ed1ddb0398eb134e6813ddde6c015'

# OpenDART에서 Zipfile 받아와 객체에 저장하기
u = requests.get('https://opendart.fss.or.kr/api/corpCode.xml', params={'crtfc_key':api_key})
zipfile_bytes = u.content
zipfile_obj = ZipFile(BytesIO(zipfile_bytes))

# 압축을 풀어서 XML File을 string으로 담기
xmlfile_objs = {name: zipfile_obj.read(name) for name in zipfile_obj.namelist()}
xml_str = xmlfile_objs['CORPCODE.xml'].decode('utf-8')

# XML String을 가져와서 DataFrame에 담기
xroot = et.fromstring(xml_str)

df_cols = ["corp_code", "corp_name", "stock_code", "modify_date"]
rows = []

for node in xroot:
    res = []
    for el in df_cols[0:]:
        if node is not None and node.find(el) is not None:
            res.append(node.find(el).text)
        else:
            res.append(None)
    rows.append({df_cols[i]: res[i]
                  for i, _ in enumerate(df_cols)})

df = pd.DataFrame(rows, columns=df_cols)
print(df)
```

```python
import pandas as pd

# Assuming df_bs and df_ci are your DataFrames

# Extracting values using iloc and converting them to numeric
ar = pd.to_numeric(df_bs.iloc[3, df_bs.columns.get_loc('20221231')], errors='coerce')
inven = pd.to_numeric(df_bs.iloc[4, df_bs.columns.get_loc('20221231')], errors='coerce')
ap = pd.to_numeric(df_bs.iloc[19, df_bs.columns.get_loc('20221231')], errors='coerce')

goods = pd.to_numeric(df_ci.iloc[1, df_ci.columns.get_loc('20220101-20221231')], errors='coerce')
sales = pd.to_numeric(df_ci.iloc[0, df_ci.columns.get_loc('20220101-20221231')], errors='coerce')

# Ensuring each value is a scalar
ar = ar.item() if isinstance(ar, pd.Series) else ar
inven = inven.item() if isinstance(inven, pd.Series) else inven
ap = ap.item() if isinstance(ap, pd.Series) else ap
goods = goods.item() if isinstance(goods, pd.Series) else goods
sales = sales.item() if isinstance(sales, pd.Series) else sales

# Calculating the ratios
inven_goods_ratio = inven / goods if goods != 0 else 0
ar_sales_ratio = ar / sales if sales != 0 else 0
ap_goods_ratio = ap / goods if goods != 0 else 0

# Perform the calculation
result = inven_goods_ratio + ar_sales_ratio - ap_goods_ratio

# Displaying the result
print("1년 데이터 기반, 현금 전환 주기 (CCC):", result)

```

```python
# 유동비율 (Quick Ratio) : Current Assets - Inventory / Current Liabilities ( 유동자산 - 재고자산/유동부채 )
```

```python
# Convert current assets to numeric and extract value
ca = pd.to_numeric(df_bs.iloc[0, df_bs.columns.get_loc('20221231')], errors='coerce')
ca_form = ca.item() if isinstance(ca, pd.Series) else ca

# Convert inventory to numeric and extract value
inv = pd.to_numeric(df_bs.iloc[4, df_bs.columns.get_loc('20221231')], errors='coerce')
inv_form = inv.item() if isinstance(inv, pd.Series) else inv

# Convert current liabilities to numeric and extract value
cl = pd.to_numeric(df_bs.iloc[18, df_bs.columns.get_loc('20221231')], errors='coerce')
cl_form = cl.item() if isinstance(cl, pd.Series) else cl

# Perform the calculation (ca_form - inv_form) / cl_form
qr = (ca_form - inv_form) / cl_form if cl_form != 0 else 0  # Check to avoid division by zero

qr
```
